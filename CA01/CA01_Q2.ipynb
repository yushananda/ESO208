```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
import math
def poly(a, g, x):

    result = a[0]
    for i in range(1, g+1):
 
        result = result*x + a[i]
  
    return result
def muller(f, x0, x1, x2, max_iter, max_err):
    itern = 0
    err = max_err*2
    while(itern<max_iter and err>max_err):
        x2_prev = x2
        h0 = x1 - x0
        h1 = x2 - x1
        del0 = (f(x1)-f(x0))/h0
        del1 = (f(x2)-f(x1))/h1
        a = (del1 - del0)/(h0+h1)
        b = a*h1 + del1
        c = f(x2)
        if(b<0):
            x2 = x2 - 2*c/(b - math.sqrt(pow(b,2)-4*a*c))
        else:
            x2 = x2 - 2*c/(b + math.sqrt(pow(b,2)-4*a*c))
        itern = itern + 1
        err = abs(x2-x2_prev)/x2 * 100
        x0 = x1
        x1 = x2_prev
        if(err<max_err or itern>max_iter):
            break
    x1 = np.linspace(-5, 5, 100)
    y1 = f(x1)
    plt.plot(x1, y1)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title("Plot of f(x) vs x")
    plt.show()
    return x2
def bairstow(a,r,s,g,roots,max_err):
    if(g<1):
        return None
    if((g==1) and (a[1]!=0)):
        roots.append(float(a[1])/float(-a[0]))
        return None
    if(g==2):
        D = (a[1]**2.0)-(4.0)*(a[2])*(a[0])
        X1 = (-a[1] - math.sqrt(D))/(2.0*a[2])
        X2 = (-a[1] + math.sqrt(D))/(2.0*a[2])
        roots.append(1/X1)
        roots.append(1/X2)
        return None
    n = g + 1
    b = [0]*n
    c = [0]*n
    b[n-1] = a[n-1]
    b[n-2] = a[n-2] + r*b[n-1]
    i = n - 3
    while(i>=0):
        b[i] = a[i] + r*b[i+1] + s*b[i+2]
        i = i - 1
    c[n-1] = b[n-1]
    c[n-2] = b[n-2] + r*c[n-1]
    i = n - 3
    while(i>=0):
        c[i] = b[i] + r*c[i+1] + s*c[i+2]
        i = i - 1
    Din = ((c[2]*c[2])-(c[3]*c[1]))**(-1.0)
    r = r + (Din)*((c[2])*(-b[1])+(-c[3])*(-b[0]))
    s = s + (Din)*((-c[1])*(-b[1])+(c[2])*(-b[0]))
    if(abs(b[0])*100>max_err or abs(b[1])*100>max_err):
        return bairstow(a,r,s,g,roots,max_err)
    if (g>=3):
        Dis = ((-r)**(2.0))-((4.0)*(1.0)*(-s))
        if(Dis<0):
            X1 = complex(r/(2.0), -(math.sqrt(-1*Dis)))
            X2 = complex(r/(2.0), (math.sqrt(-1*Dis)))
        if(Dis>0):
            X1 = (r - (math.sqrt(Dis)))/(2.0)
            X2 = (r + (math.sqrt(Dis)))/(2.0)
        roots.append(1/X1)
        roots.append(1/X2)
        return bairstow(b[2:],r,s,g-2,roots, max_err)
print("1. Muller's Method")
print("2. Bairstow's Method")
t = int(input("Enter the serial number of one of the above methods to use them: "))
if(t==1):    
    func = input("Enter the function f(x) = ")
    x = Symbol('x')
    f = lambdify(x, func)
    x0 = float(input("Enter first initial value: "))
    x1 = float(input("Enter second initial value: "))
    x2 = float(input("Enter third initial value: "))
    max_iter = int(input("Enter maximum number of iterations: "))
    max_err = float(input("Enter maximum permissible error: "))
    mull = muller(f, x0, x1, x2, max_iter, max_err)
    print("Approximate solution of the equation is", mull)
if(t==2):
    a = []
    r = float(input("Enter first initial value: "))
    s = float(input("Enter second initial value: "))
    #max_iter = int(input("Enter maximum number of iterations: "))
    max_err = float(input("Enter maximum permissible error: "))
    g = int(input("Enter the degree of the polynomial: "))
    print("Enter the coefficients in decreasing order: ")
    for i in range (0, g+1):
        ele = float(input())
        a.append(ele)
    x1 = np.linspace(-5,5,100)
    y1 = poly(a,g,x1)
    plt.plot(x1, y1)
    plt.show()
    roots = []
    bairstow(a,r,s,g,roots,max_err)
    print("The roots (both complex and real) are: ")
    for r in roots:
        print(str(r) + " ")
```
